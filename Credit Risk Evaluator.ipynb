```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.metrics import classification_report
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data

train_cols = train_df.columns
i=0
for col in train_cols:
    print (col)
    i=i+1
print("total cols:", i)
```

```python
# Convert categorical data to numeric and separate target feature for testing data
test_df=pd.get_dummies(test_df)
test_cols= test_df.columns
for col in train_cols:
    print (col)
    i=i+1
print("total cols:", i)
```

```python
y_train= train_df['loan_status_low_risk']
X_train= train_df.drop(['loan_status_low_risk'],axis=1)
y_test= test_df['loan_status_low_risk']
X_test= test_df.drop(['loan_status_low_risk'], axis=1)
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg
logreg.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {logreg.score(X_train, y_train)}")
print(f"Testing Data Score: {logreg.score(X_test,y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
ranfor = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)
print(f'Training Score: {ranfor.score(X_train, y_train)}')
print(f'Testing Score: {ranfor.score(X_test, y_test)}')
```

# Conclusion: What we can see that the random forest model fits better in comparison with the logistic regression model

```python
# Scale the data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled.shape
#X_test_scaled.shape
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
logreg_scaled = LogisticRegression()
# classifier
logreg_scaled.fit(X_train_scaled, y_train)
y_pred = logreg_scaled.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
ranfor_scaled = RandomForestClassifier(random_state=1).fit(X_train_scaled, y_train)
ranfor_scaled.fit(X_train_scaled, y_train)
y_pred = ranfor_scaled.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
print(f"Training Data Score: {ranfor_scaled.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {ranfor_scaled.score(X_test_scaled, y_test)}")
```

# Conclusion: After scaling  random forest model as well as logistic regression model fit well
